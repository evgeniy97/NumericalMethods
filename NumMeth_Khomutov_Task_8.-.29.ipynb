## Задание 1


```python
c = np.array([2.5,-3.3,0,8.4,-5.2,0.9])

a = 1
b = 1.44

epselon = 10**-4

def P(x):
    s = 0
    for i in range(6):
        s += c[i] * x**i
    return s
```

$$

```python
def fff(x):
    return 20*0.9*x**3-12*5.2*x
```
