## Задание 1

Определим начальные данные задачи

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.misc import derivative
%matplotlib inline
```

```python
def f(x):
    return 2* np.sin(x) - 3*np.cos(x)
A = -1
B = 1
epselon = 10**(-6)
```

Построим график функции $f(x)$ на рассматриваемом промежутке:

```python
X = np.arange(A, B, 0.01)

fig = plt.figure() 
plt.plot(X, f(X))
plt.show()
```

Из графика видно, что минимум функции достигается около точки -0.5, а максимум - на правом конце промежутка.
Если производная на рассматриваемом отрезке будет иметь только один корень, то наше предположение верно.

Производная $\frac{df(x)}{dx} = 2 cos(x) + 3 sin(x)$

```python
def df(x): 
    return 2* np.cos(x) + 3*np.sin(x)
```

График производной:

```python
fig = plt.figure() 
plt.plot(X, df(X))
plt.show()
```

Наше предположение оказалось верным, следовательно максимум функции $f(x)$ достигается в точке 1:

```python
Maxf = f(1)
```

Определим метод Ньютона и применим его к производной функции $f(x)$, и найдем ее ноль (ноль производной):

```python
def NewtoonMethod(f,x,epselon):
    n = 1
    x_old = x
    x_new = x_old - f(x_old)/derivative(f,x_old,dx = epselon)
    while abs(x_new - x_old) >= epselon:
        x_old = x_new
        x_new = x_old - f(x_old)/derivative(f,x_old,dx = epselon)
        n = n + 1
    return x_new, n
```

```python
xmin, n = NewtoonMethod(df,-0.5,epselon)
print(xmin,n)
```

```python
Minf = f(xmin)
```

```python
print('Max:',Maxf)
print('Min:',Minf)
```

Следовательно,
$$
max(f(x)) = 0.0620350520114 \\
min(f(x)) = -3.60555127546
$$

```python
a = 0
b = 5
def g(t):
    return t*np.sin(t) 
```

Построим график исходной функции:

```python
X = np.arange(a, b, 0.01)
fig = plt.figure() 
plt.plot(X, g(X))
plt.show()
```

Функция для вычисления чисел Фибоначчи:

```python
M = {0: 0, 1: 1}

def fib(n):
    if n in M:
        return M[n]
    M[n] = fib(n - 1) + fib(n - 2)
    return M[n]
```

Определим метод Фибоначчи:

```python
def Fibonacci(g,a,b,epselon):
    
    n = 1;
    while (b-a)/fib(n) > epselon:
        n=n+1
    print(n)
    while n != 1:
        lamda = a + (b - a)*(fib(n-2)/fib(n))
        mu = a + (b - a)*(fib(n-1)/fib(n))

        if g(lamda) <= g(mu):
            b = mu
            x = lamda
        else:
            a = lamda
            x = mu

        n -= 1

        if n == 1:
            return((mu + lamda) / 2)
    return x
```

Метод Фибоначчи работает для нахожданения минимума, поэтому определим функцию $-g(x)$:

```python
def gminus(x):
    return -g(x)
```

Найдем точку максимума функции. Из графика видо, что эта точка лежит в промежутке [1;3]:

```python
xmax = Fibonacci(gminus,1,3,epselon)
xmax
```

Найдем точку максимума функции. Из графика - точка лежит в промежутке [4;5]:

```python
xmin = Fibonacci(g,4,5,epselon)
xmin
```

```python
print("Max: ",g(xmax))
print("MIn: ",g(xmin))
```
