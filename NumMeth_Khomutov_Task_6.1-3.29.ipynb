## Задание №2

Функция имеет вид
$$
x(t) = \nu t + b
$$

Можно всю эту красоту представить как скалярное произведение  расширенного вектора признаков (термины из машинного обучения) $t = (1,t)$ на вектор коэфицентов  $ w = (\nu,b) $. Тогда $x(t)$ - целевой признак.

Тогда вектор коэфицентов можно найти как 
$$
w = (t^T t)^{-1} t^T x
$$

```python
t = np.array([2,3.2,4.4,5,5.6,6.8,7.4,8])
x = np.array([18.5,35.73,54.65,62.4,71.4,90.5,98.1,107.6])
```

```python
tt = t.reshape(t.size,1)
tt = np.insert(tt, 0, 1, axis=1)

xx = x.reshape(x.size,1)
```

```python
w = np.linalg.inv(np.dot(tt.T,tt)).dot(tt.T).dot(xx)
print(w)
```

Следовательно наша искомая функция имеет вид:
$$
x(t) = 14.87142857t - 11.45857143
$$

```python
def X(t):
    return w[1]*t + w[0]
```

```python
T = np.arange(2, 8, 0.01)

fig = plt.figure() 
plt.plot(T, X(T))
plt.scatter(t,x,color ='r')
plt.show() 
print("Прогноз на 2T: ",X(16))
```

## Задание №3


Сведем задачу к линейно зависимости:
$$
y = (ax + b)sin(x) \\
\frac{y}{sin(x)} = ax+b \\
Пусть\: s = \frac{y}{sin(x)}, тогда: \\
s = ax + b
$$
Полученная задача решается таким же методом, как и задача №2

```python
x_ = np.arange(0.5, 3.25, 0.25)
y_ = np.array([1.7499,2.5732,3.2817,3.8197,4.1396,4.2065,3.5208,2.7829,1.8224,0.6915,0.6915])
```

```python
s = np.array([ y_[i]/np.sin(x_[i]) for i in range(0, len(x_))])
s = s.reshape(11,1)
```

```python
x__ = x_.reshape(11,1)
x__ = np.insert(x__, 0, 1, axis=1)
```

$$
w = (x^T x)^{-1} x^T s
$$

```python
w_ = np.linalg.inv(np.dot(x__.T,x__)).dot(x__.T).dot(s)
print(w_)
```

```python
def y(x):
    return (w_[1] * x + w_[0])*np.sin(x)
```

```python
T = np.arange(0.5, 3, 0.01)
fig = plt.figure() 
plt.plot(T, y(T))
plt.scatter(x_,y_,color ='r')
plt.show() 
```
